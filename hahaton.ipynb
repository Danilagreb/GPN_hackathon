```python
list_cities = df_merge['city'].unique() # список городов для дальнейшего отбора городов, в которых проводятся акции
list_cities
```

```python
df_merge_pos = df_merge[df_merge['neg_sale'] < df_merge['pos_sale']] # датафрейм с позитивными значениям
df_merge_pos
# соотношение выгоды по отношению к убыткам от негативных групп
# если литры, принесшие убыток (негативные группы не платят за выпивку), меньше, чем литры, принесшие прибыль, то считаем,
# что акция принесла выгоду
```

```python
df_merge_pos['diff_pos_neg'] = df_merge_pos['pos_sale'] - df_merge_pos['neg_sale']
df_merge_pos
# посчитаем разницу между выгодой и убытком в литрах, чтобы учитывать прибыль в литрах
```

```python
list_effects = promo['Название'].to_list()
list_effects
```

```python
list_ef = promo['Название'].unique()
```

```python
zero = [0] * 49
```

```python
df = pd.DataFrame({'Акция': list_ef, 'Объем напитков':zero}) # будет итоговый датафрэйм с данными по акциям и объемы выручки в литрах
df
```

```python
df_merge_pos
```

```python
# алгоритм для высчитывания значений

for elem in list_effects:
    cities = [] # города, в которых проходит определенная акция
    sep_df = promo.iloc[[promo[promo['Название'] == elem].index[0]]] # датафрейм для отбора городов, участвующих в акции
    for cit in list_cities: # отбор городов, в которых проводится акция
        if sep_df.at[sep_df[sep_df['Название'] == elem].index[0], cit] == 1:
            cities.append(cit)
    
    
    
    alco = promo.at[promo[promo['Название'] == elem].index[0], 'Алкогольное'] # флаг на акцию алкогольного
    no_alco = promo.at[promo[promo['Название'] == elem].index[0], 'Безалкогольное'] # флаг на акцию безалкогольного
    start = promo.at[promo[promo['Название'] == elem].index[0], 'start_ts'] # начало акции
    end = promo.at[promo[promo['Название'] == elem].index[0], 'end_ts'] # конец акции
    # итоговая таблица, из которой выбираем данные по акции для суммирования
    if alco == 1:
        df_elem_alc = df_merge_pos.query("date_ts >= @start and date_ts <= @end and product == 'Алкогольные' and city in @cities")
        df.loc[df['Акция'] == elem, 'Объем напитков'] += df_elem_alc['diff_pos_neg'].sum()
    if no_alco == 1:
        df_elem_noalc = df_merge_pos.query("date_ts >= @start and date_ts <= @end and product == 'Безалкогольные' and city in @cities")
        df.loc[df['Акция'] == elem, 'Объем напитков'] += df_elem_noalc['diff_pos_neg'].sum()
df # получаем итоговый датафрейм с объемом выгоды у каждой акции
```

```python
df.to_csv('promo_effects.csv') # итоговый файл с принесенной прибылью от акций
```
